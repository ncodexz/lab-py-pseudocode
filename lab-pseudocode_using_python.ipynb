### Exercise 1: Multiply Two Numbers

Write a pseudocode to read two numbers and multiply them together. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->
**Pseudocode:**
    // Declare variables
    number1, number2, result : Real

### Exercise 2: Check Even or Odd

Write pseudocode that checks if a number is even or odd and prints the result. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->
**Pseudocode:**
```
START
    // Read a number
    Write "Enter a number: "
    Read number3
    // Check if the number is even or odd
    IF number3 divided by 2 or remainder is 0 THEN
        display "The number is even."
    ELSE
        display "The number is odd."
END
```
     

````python
# Add your code solution here
# Read the third number
number3 = float(input("Enter the third number: "))

# Check if the number is even or odd
if number3 % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
````

### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->
**Pseudocode:**
```
START
    //Read a natural number
    Write "Enter a natural number:"
    Read Natural_number
    // Set sumfirst_n to 0
    sumfirst_n = 0
    FOR i from 1 to Natural_number DO
        sumfirst_n = sumfirst_n + i
    END FOR
    // Display the result
    Write "The sum of the first ", Natural_number, " natural numbers is: ", sumfirst_n
END
```

````python
# Add your code solution here

# Read a natural number
natural_number = int(input("Enter a natural number: "))

# Set sumfirst_n to 0
sumfirst_n = 0
for i in range(1, natural_number + 1):
    sumfirst_n += i

# Display the result
print("The sum of the first", natural_number, "natural numbers is:", sumfirst_n)
````

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->
**Pseudocode:**
```
START
    // Read imput string
    Write "Enter a string: "
    Read input_string
    // Initialize vowel count
    vowel_count = 0
    // Loop through each character in the string
    FOR each character in input_string DO
        IF Character is in "aeiouAEIOU" THEN
            vowel_count = vowel_count + 1
        END IF
    END FOR
    // Display the result
    Write "The number of vowels in the string is: ", vowel_count
END

````python
# Add your code solution here
input_string = input("Enter a string: ")
vowel_count = 0
for char in input_string:
    if char in "aeiouAEIOU":
        vowel_count += 1
print("The number of vowels in the string is:", vowel_count)
````

### Bonus Exercise 9: Find Factorial of a Number

```
START
